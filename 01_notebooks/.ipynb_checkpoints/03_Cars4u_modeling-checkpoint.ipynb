###### Data Transformation

```python
skewed = data[numeric_cols].apply('skew')
skewed[skewed > 2.0].index
```

```python
def log_transform(data, cols):
    for colname in cols:
        if (data[colname] <= 0).any():
            min_val = data[colname].min()
            if min_val < 0:
                adjustment = abs(min_val) + 1
                data[colname + '_log'] = np.log(data[colname] + adjustment)
            else:
                data[colname + '_log'] = np.log(data[colname] + 1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
```

```python
log_transform(data, skewed[skewed > 2.0].index)
```

```python
def distplot(df, columns):
    for col in columns:
        print(col)
        print(f'Skew: {round(df[col].skew(), 2)}')
        plt.figure(figsize = (15, 4))
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True, bins=20)
        plt.ylabel('Frequence')
        plt.xlabel(col)
        plt.title(f'Distribution of {col}')
        plt.subplot(1, 2, 2)
        sns.boxplot(x=data[col])
        plt.title(f'Box plot of {col}')
        plt.show()
```

```python
from sklearn.model_selection import train_test_split
#X = df_data[['Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage_kmpl', 'Engine_CC', 'Power_bhp',
#       'Seats', 'Car_Age', 'Kilometers_Driven_log', 'avg_price_log_brand_model']]
X = df_data[['Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage_kmpl', 'Engine_CC', 'Power_bhp',
       'Seats', 'Car_Age', 'Kilometers_Driven_log']]
y = df_data[['Price_log']]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import itertools

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['Location', 
                                                                   'Fuel_Type', 
                                                                   'Transmission', 
                                                                   'Owner_Type',])], remainder='passthrough')

feature_columns = np.concat([
    np.char.add('Location_',  np.sort(X['Location'].unique())),
    np.char.add('Fuel_Type_',  np.sort(X['Fuel_Type'].unique())),
    np.char.add('Transmission_',  np.sort(X['Transmission'].unique())),
    np.char.add('Owner_Type_',  np.sort(X['Owner_Type'].unique())),
    X.columns[4:]
])
X = pd.DataFrame(ct.fit_transform(X))
X.columns = feature_columns

X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12345)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
def R2(X, y, pred):
    R_2 = r2_score(y, pred)
    R_2_adj = 1 - (1 - R_2)*((X.shape[0] - 1)/(X.shape[0] - len(X.columns) - 1))
    return (R_2, R_2_adj)
```

```python
r_2, r_2_adj = R2(X_train, y_train, y_pred)
print(f'R^2 = {r_2}')
print(f'R^2 adj. = {r_2_adj}')
```

```python
train_df = pd.DataFrame()
train_df['Actual Price_log'] = y_train.iloc[:, 0]
train_df['Predicted Price_log'] = y_pred.reshape((1, len(y_pred)))[0]

train_df.head()
```

```python
MSE = mean_squared_error(train_df['Actual Price_log'], train_df['Predicted Price_log'])
print(f'MSE = {MSE}')
```

```python
# test model
r_2, r_2_adj = R2(X_test, y_test, model.predict(X_test))
print(f'R^2 = {r_2}')
print(f'R^2 adj. = {r_2_adj}')
```

```python
MSE = mean_squared_error(y_test, model.predict(X_test))
print(f'MSE = {MSE}')
```
