Unit __km/kg__ is specific for CNG and LPG fuel types, __kmpl__ is for disel/petrol fuel type. Let's convert `Mileage_num` to common unit __kmpl__. According to [Units & Conversations Fact Sheet](https://cngcenter.com/wp-content/uploads/2013/09/UnitsAndConversions.pdf) by MIT Energy Club CNG @ 20MPa $\approx$ 0.185 kg/L and LPG $\approx$ 0.540 kg/L.

```python
gas_density_kgpl = {'LPG': 0.540, 'CNG': 0.185}
def convert_mileage_unit(row):
    """
        Converts mileage for gas fuel from km/kg to kmpl.
    """
    if row['Fuel_Type'] in ['CNG', 'LPG']:
        return row['Mileage_num'] * gas_density_kgpl[row['Fuel_Type']]
    else:
        return row['Mileage_num']
```

```python
data['Mileage_kmpl'] = data.apply(convert_mileage_unit, axis=1)
data.head()
```

## Rename columns <a name='rename-columns'></a>

```python
# rename columns to snake_case
def clean_columns(df, replace={}):
    """
    Clean column names of a DataFrame based on a replacement dictionary and standardize them.
    
    Parameters:
    df (pd.DataFrame): DataFrame whose columns need to be cleaned.
    replace (dict): Dictionary where keys are substrings to find and values are substrings to replace them with.
    
    Returns:
    pd.DataFrame: DataFrame with cleaned and standardized column names.
    """
    columns = df.columns.copy()
    for old, new in replace.items():
        columns = columns.str.replace(old, new, regex=False)

    columns = columns.str.lower()
    columns = columns.str.replace(' ', '_')
    columns = columns.str.replace(r'[^\w\s]', '', regex=True)
    columns = columns.str.strip()
    df.columns = columns
    
    return df

# Dictionary of replacements (example)
replace_dict = {
    '/': ' or ', # TODO: replace...
}

# Clean column names
data = clean_columns(data, replace=replace_dict)

data.columns
```

```python
def distplots(df, columns):
    for col in columns:
        distplot(df[col], col)
```

```python
def distplot(var, var_name):
    print(f'`{var_name}`')
    print(f'Skew: {round(var.skew(), 2)}')
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(var, kde=True, bins=20)
    plt.ylabel('Frequence')
    plt.xlabel(var_name)
    plt.title(f'Distribution of {var_name}')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=var)
    plt.title(f'Box plot of {var_name}')
    plt.show()
```

```python
# drop unlogical predictors for new_car_price
new_data = data.drop(columns=['kilometers_driven', 'owner_type', 'price_lakh'])
```

```python
new_data.columns
```

```python
new_data = pd.get_dummies(new_data, columns=['location', 'fuel_type', 'transmission', 'brand', 'model'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
# split data without missing `new_car_price` to train and valicdation set
train_data, test_data = train_test_split(new_data[new_data['new_car_price_lakh'].notna()],
                                   test_size=0.3, random_state=12345)
```

```python
X_train = train_data.drop(columns=['new_car_price_lakh'])
y_train = np.log(train_data['new_car_price_lakh']) # log it because the variable is right skewed
X_test = test_data.drop(columns=['new_car_price_lakh'])
y_test = np.log(test_data['new_car_price_lakh']) # log it because the variable is right skewed
```

```python
from sklearn.linear_model import LinearRegression
impute_model = LinearRegression()
impute_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
def model_performance(model, X, y, name):
    # predict
    predicted = model.predict(X)
    print(f'R^2 = {r2_score(y, predicted)}')
    print(f'MAE = {mean_absolute_error(y, predicted)}')
    print(f'MAPE = {mean_absolute_percentage_error(y, predicted)*100}%')
    
    plt.scatter(y, predicted)
    plt.plot([min(y), max(y)], \
             [min(y), max(y)], color='red', linestyle='--')
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title(f"Actual vs Predicted on {name} set")
    plt.show()
```

```python
model_performance(impute_model, X_train, y_train, 'Train')
```

```python
model_performance(impute_model, X_test, y_test, 'Test')
```

```python
# impute missing `new_car_price_lakh`

data.loc[data['new_car_price_lakh'].isna(), 'new_car_price_lakh'] =  \
        np.exp(impute_model.predict(new_data[new_data['new_car_price_lakh'].isna()].drop(columns=['new_car_price_lakh'])))
```
